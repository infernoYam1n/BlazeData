````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import gridspec
````

------

````python
# üì¶ 1. Sample Dataset Create
np.random.seed(42)
n = 100
df = pd.DataFrame({
    'age': np.random.randint(18, 60, size=n),
    'income': np.random.normal(50000, 12000, size=n),
    'gender': np.random.choice(['Male', 'Female'], size=n),
    'score': np.random.uniform(0, 100, size=n),
    'group': np.random.choice(['A', 'B', 'C'], size=n)
})

# üìä 2. All Plots Together
plt.figure(figsize=(20, 50))

# 1. Scatter Plot
plt.subplot(9, 2, 1)
plt.scatter(df['age'], df['income'], alpha=0.6)
plt.title('Scatter Plot (Age vs Income)')

# 2. Line Plot
plt.subplot(9, 2, 2)
plt.plot(sorted(df['age']), sorted(df['score']))
plt.title('Line Plot (Age vs Score)')

# 3. Step Plot
plt.subplot(9, 2, 3)
plt.step(sorted(df['age']), sorted(df['income']))
plt.title('Step Plot')

# 4. Histogram
plt.subplot(9, 2, 4)
plt.hist(df['income'], bins=15, color='skyblue')
plt.title('Histogram (Income)')

# 5. Bar Chart (Categorical)
plt.subplot(9, 2, 5)
group_means = df.groupby('group')['score'].mean()
plt.bar(group_means.index, group_means.values, color='orange')
plt.title('Bar Chart (Avg Score per Group)')

# 6. Horizontal Bar
plt.subplot(9, 2, 6)
plt.barh(group_means.index, group_means.values, color='green')
plt.title('Horizontal Bar Chart')

# 7. Pie Chart
plt.subplot(9, 2, 7)
gender_counts = df['gender'].value_counts()
plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart (Gender)')

# 8. Box Plot
plt.subplot(9, 2, 8)
plt.boxplot(df['income'])
plt.title('Box Plot (Income)')

# 9. Heatmap (Manual using imshow)
plt.subplot(9, 2, 9)
corr = df[['age', 'income', 'score']].corr()
plt.imshow(corr, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
plt.title('Heatmap (Correlation)')

# 10. Contour Plot
plt.subplot(9, 2, 10)
X, Y = np.meshgrid(np.linspace(-3,3,100), np.linspace(-3,3,100))
Z = np.sin(X**2 + Y**2)
plt.contour(X, Y, Z, cmap='plasma')
plt.title('Contour Plot')

# 11. Subplot Example
plt.subplot(9, 2, 11)
plt.plot(df['age'], df['score'])
plt.title('Subplot Example (1 of many)')

# 12. Stack Plot
plt.subplot(9, 2, 12)
days = np.arange(1, 6)
product1 = [3, 4, 5, 6, 7]
product2 = [1, 2, 3, 2, 1]
plt.stackplot(days, product1, product2, labels=['Prod1', 'Prod2'])
plt.legend()
plt.title('Stack Plot')

# 13. Error Bar Plot
plt.subplot(9, 2, 13)
x = np.arange(5)
y = np.random.normal(5, 1, size=5)
errors = np.random.rand(5) * 0.5
plt.errorbar(x, y, yerr=errors, fmt='-o')
plt.title('Error Bar Plot')

# 14. Fill Between
plt.subplot(9, 2, 14)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.fill_between(x, y - 0.2, y + 0.2, alpha=0.3)
plt.title('Fill Between Plot')

# 15. Annotate
plt.subplot(9, 2, 15)
plt.plot(df['age'], df['score'], 'o')
plt.annotate('Young High Scorer', xy=(df['age'][0], df['score'][0]), xytext=(30, 90),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.title('Annotation Example')

# 16. Twin Axes
plt.subplot(9, 2, 16)
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(df['age'], df['score'], 'g-')
ax2.plot(df['age'], df['income'], 'b--')
ax1.set_xlabel('Age')
ax1.set_ylabel('Score', color='g')
ax2.set_ylabel('Income', color='b')
plt.title('Twin Axes Plot')
plt.tight_layout()

plt.show()
````

````python
df=pd.read_csv(r'E:\DataBlaze\Public-Datasets-main\house.csv')

#Basic information about the dataset
print(df.info())
print(df.head())

#check for missing values
print(df.isnull().sum())

#Quick statistical summary
print(df.describe())
````

````python
num_cols = ['SalePrice', 'LotArea', 'GrLivArea', 'YearBuilt']
cat_cols = ['Street', 'Neighborhood', 'HouseStyle', 'OverallQual']

# 1. Histogram
for col in num_cols:
    plt.hist(df[col], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

# 2. Bar Chart (Categorical Count)
for col in cat_cols:
    df[col].value_counts().plot(kind='bar', color='orange')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 3. Pie Chart
df['Street'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Street Type Distribution')
plt.ylabel('')
plt.show()

# 4. Box Plot
for col in num_cols:
    plt.boxplot(df[col].dropna())
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.grid(True)
    plt.show()

# 5. Scatter Plot (Numerical Relationship)
plt.scatter(df['GrLivArea'], df['SalePrice'], alpha=0.5)
plt.title('GrLivArea vs SalePrice')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.grid(True)
plt.show()

# 6. Line Plot (Trend)
df_sorted = df.sort_values('YearBuilt')
plt.plot(df_sorted['YearBuilt'], df_sorted['SalePrice'], color='green')
plt.title('YearBuilt vs SalePrice')
plt.xlabel('YearBuilt')
plt.ylabel('SalePrice')
plt.grid(True)
plt.show()

# 7. Correlation Heatmap (Matplotlib)
corr = df[num_cols].corr()
fig, ax = plt.subplots()
cax = ax.matshow(corr, cmap='coolwarm')
fig.colorbar(cax)
plt.xticks(range(len(num_cols)), num_cols, rotation=90)
plt.yticks(range(len(num_cols)), num_cols)
plt.title('Correlation Heatmap (Matplotlib)', pad=20)
plt.show()

# 8. Multiple Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].hist(df['SalePrice'], bins=30, color='skyblue')
axs[0, 0].set_title('SalePrice Histogram')

axs[0, 1].boxplot(df['LotArea'].dropna())
axs[0, 1].set_title('LotArea Boxplot')

axs[1, 0].scatter(df['GrLivArea'], df['SalePrice'], alpha=0.5)
axs[1, 0].set_title('GrLivArea vs SalePrice')

df['Street'].value_counts().plot(kind='bar', ax=axs[1, 1], color='orange')
axs[1, 1].set_title('Street Bar Chart')

plt.tight_layout()
plt.show()

# 9. Object-Oriented (OO) Plotting Interface
fig, ax = plt.subplots()
ax.plot(df['SalePrice'], color='blue')
ax.set_title("OO Interface Plot")
ax.set_xlabel("Index")
ax.set_ylabel("SalePrice")
plt.show()

# 10. Annotation and Legend
plt.plot(df['GrLivArea'], df['SalePrice'], label='Price Line', color='purple')
plt.title("Annotation & Legend")
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.legend()
plt.annotate('High Price', xy=(3000, 600000), xytext=(3200, 650000),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()

# 11. Shared Axis Subplots
fig, axs = plt.subplots(1, 2, sharey=True)
axs[0].hist(df['LotArea'], color='teal')
axs[0].set_title("LotArea")
axs[1].hist(df['GrLivArea'], color='coral')
axs[1].set_title("GrLivArea")
plt.show()

# 12. Dual Axes Plot
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(df['GrLivArea'], df['SalePrice'], 'g-', label='SalePrice')
ax2.plot(df['GrLivArea'], df['LotArea'], 'b-', label='LotArea')
ax1.set_xlabel('GrLivArea')
ax1.set_ylabel('SalePrice', color='g')
ax2.set_ylabel('LotArea', color='b')
plt.title("Dual Y-Axis Plot")
plt.show()

# 13. Color Mapping (Scatter with colormap)
plt.scatter(df['GrLivArea'], df['SalePrice'], c=df['LotArea'], cmap='viridis')
plt.title("Colored by LotArea")
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.colorbar(label='LotArea')
plt.show()

# 14. 3D Scatter Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['GrLivArea'], df['LotArea'], df['SalePrice'], c='red')
ax.set_xlabel('GrLivArea')
ax.set_ylabel('LotArea')
ax.set_zlabel('SalePrice')
plt.title("3D Scatter Plot")
plt.show()

# 15. Gridspec Layout
fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(2, 2)
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(df['SalePrice'], color='navy')
ax1.set_title('Top Full Width')

ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(df['LotArea'], color='gold')
ax2.set_title('Bottom Left')

ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(df['GrLivArea'], df['SalePrice'], alpha=0.6)
ax3.set_title('Bottom Right')

plt.tight_layout()
plt.show()

# 16. Apply Style
plt.style.use('ggplot')  # Other styles: 'seaborn', 'bmh', 'dark_background'
plt.plot(df['SalePrice'])
plt.title('Styled Plot (ggplot)')
plt.show()

# 17. Save Figure
plt.plot(df['LotArea'], df['SalePrice'])
plt.title("LotArea vs SalePrice")
plt.savefig("lotarea_vs_saleprice.png", dpi=300, bbox_inches='tight')
plt.show()

# 18. Customizing Ticks and Labels
plt.plot(df['SalePrice'])
plt.title("Custom Ticks and Labels")
plt.xticks(ticks=[0, 100, 200, 300, 400], labels=['zero', 'one hundred', 'two hundred', 'three hundred', 'four hundred'])
plt.yticks(np.arange(0, df['SalePrice'].max(), 100000))
plt.grid(True)
plt.show()

# 19. Log Scale Plot
plt.scatter(df['GrLivArea'], df['SalePrice'])
plt.xscale('log')
plt.yscale('log')
plt.title("Log-Log Scale Scatter Plot")
plt.xlabel("GrLivArea (log scale)")
plt.ylabel("SalePrice (log scale)")
plt.grid(True)
plt.show()

# 20. Filling Between Curves
x = np.linspace(0, 100, 200)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5
plt.fill_between(x, y1, y2, color='lightblue')
plt.title("Fill Between Curves")
plt.show()

# 21. Step Plot
plt.step(range(10), np.random.randint(1, 10, 10))
plt.title("Step Plot")
plt.show()

# 22. Error Bars
x = np.arange(10)
y = np.random.rand(10) * 10
yerr = np.random.rand(10)
plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bars")
plt.show()

# 23. Histogram with cumulative distribution
plt.hist(df['SalePrice'], bins=30, cumulative=True, color='purple', edgecolor='black')
plt.title("Cumulative Histogram of SalePrice")
plt.show()
````
