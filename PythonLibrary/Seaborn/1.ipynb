
🎨 1. hue → রঙ (Color)
👉 কি করে:
একই গ্রাফে বিভিন্ন ডেটা গ্রুপকে ভিন্ন ভিন্ন রঙে দেখায়।

👉 কবে ব্যবহার করবো:
যখন তোমার ডেটাসেটে এমন কোনো column আছে যার উপর ভিত্তি করে তুমি বিভিন্ন গ্রুপ আলাদা করতে চাও।

📌 Example:

sns.relplot(data=tips, x="total_bill", y="tip", hue="sex")

🔹 এখানে sex (Male/Female) অনুযায়ী দুইটি আলাদা রঙ হবে।

🧭 2. style → চিহ্ন বা লাইন স্টাইল (Marker/Line Style)
👉 কি করে:
গ্রাফে আলাদা গ্রুপকে ভিন্ন চিহ্ন (যেমন: ○, △, ⬠) বা লাইন স্টাইল (যেমন: সরল, ড্যাশড) দিয়ে দেখায়।

👉 কবে ব্যবহার করবো:
যখন রঙ ছাড়াও আরেকটি dimension দিয়ে গ্রুপ আলাদা করে বোঝাতে চাও।

📌 Example:

sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker")

🔹 এখন sex অনুযায়ী রঙ, আর smoker (Yes/No) অনুযায়ী ভিন্ন মার্কার/লাইন স্টাইল।

⚠️ style এর জন্য categorical (শ্রেণিভুক্ত) column ব্যবহার করো। Numeric দিলে বোঝা যাবে না কোনটা কোন গ্রুপ।

📏 3. size → আকার (Size)
👉 কি করে:
ডেটা পয়েন্টের আকার (বড়/ছোট) পরিবর্তন করে একটি ভিন্ন মান বোঝায়।

👉 কবে ব্যবহার করবো:
যখন কোনো সংখ্যা (যেমন: আয়, value, weight) অনুযায়ী আকার আলাদা করতে চাও।

📌 Example:


sns.relplot(data=tips, x="total_bill", y="tip", size="size", sizes=(40, 200))

🔹 এখানে size অনুযায়ী ছোট/বড় গোলাকার চিহ্ন দেখাবে।

📈 4. kind → plot-এর ধরন
kind='scatter' → ডিফল্ট, একেকটা ডেটা পয়েন্ট।

kind='line' → পয়েন্ট গুলোকে লাইন দিয়ে যুক্ত করে।

📌 Example:

sns.relplot(data=tips, x="day", y="total_bill", kind="line")


🔹 দিনে দিনে মোট বিল কিভাবে পরিবর্তন হয়েছে সেটা দেখাবে।

<center><h2>Seaborn Data Visualization Workflow for House Dataset</h2></center>


0. Import Libraries & Set Style

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
##import anything else you need
```

```python
# Set Seaborn style and context for better visuals
sns.set(style="whitegrid", context="notebook", palette="husl")
```

1. Load & Inspect the Data

```python
df=pd.read_csv(r'E:\DataBlaze\Public-Datasets-main\house.csv')

#Basic information about the dataset
print(df.info())
print(df.head())

#check for missing values
print(df.isnull().sum())

#Quick statistical summary
print(df.describe())
```

2. Understand Data Structure

```python
print("ColumnsName :", df.columns.tolist())  # Shows the column names
print("Shape of the DataFrame:", df.shape)  # Shows the size: (rows, columns)

print("Data types:\n", df.dtypes)  # Shows what kind of data each column holds
```

### Data Types

float64 = decimal numbers (like 1.5, 2.3)

### 3. Univariate Analysis

3.1. Numerical Features (Histogram + KDE)

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['SalePrice'], kde=True, color='royalblue')
plt.title('Distribution of House Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

3.2. Categorical Features (Countplot)

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='BedroomAbvGr', data=df)
plt.title('Number of Houses by Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Count')
plt.show()
```

### 4. Bivariate Analysis

4.1. Scatter Plot (Numeric vs Numeric)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='LotArea', y='SalePrice', data=df, hue='BedroomAbvGr', palette='viridis')
plt.title('Area vs Price (Colored by Bedrooms)')
plt.xlabel('LotArea')
plt.ylabel('SalePrice')
plt.show()
```

4.2. Box Plot (Numeric vs Categorical)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='HouseStyle', y='SalePrice', data=df)
plt.title('Sale Price Distribution by House Style (Stories)')
plt.xlabel('House Style')
plt.ylabel('Sale Price')
plt.xticks(rotation=45)
plt.show()
```

### 5. Multivariate & Advanced Visualizations

5.1. Pairplot (Scatter Matrix)

```python
cols = ['SalePrice', 'GrLivArea', 'BedroomAbvGr', 'FullBath']  # Use FullBath, HalfBath if needed separately
sns.pairplot(df[cols], diag_kind='kde')
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()
```

5.2. Correlation Heatmap

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

5.3. Violin Plot (Distribution + KDE by Category)

```python
plt.figure(figsize=(8, 5))
# No furnishingstatus column, substitute with 'CentralAir' (Y/N)
sns.violinplot(x='CentralAir', y='SalePrice', data=df)
plt.title('Sale Price Distribution by Central Air Conditioning')
plt.xlabel('Central Air')
plt.ylabel('Sale Price')
plt.show()
```

5.4. Jointplot (Scatter + Distribution)

```python
sns.jointplot(x='GrLivArea', y='SalePrice', data=df, kind='reg', height=7)
plt.suptitle('Joint Distribution of Living Area and Sale Price', y=1.02)
plt.show()
```

### 6. Faceted/Grouped Visualizations

6.1. FacetGrid (Multiple Subplots by Category)

```python
g = sns.FacetGrid(df, col='CentralAir', row='GarageType', height=4)
g.map_dataframe(sns.scatterplot, x='GrLivArea', y='SalePrice')
g.add_legend()
plt.show()
```

### 7. Regression & Categorical Plots

7.1. Regression Plot

```python
sns.lmplot(x='GrLivArea', y='SalePrice', data=df, hue='CentralAir', height=6)
plt.title('Living Area vs Sale Price with Regression Lines by Air Conditioning')
plt.show()
```

7.2. Boxen Plot (Enhanced Box Plot)

```python
sns.catplot(x='HouseStyle', y='SalePrice', kind='boxen', data=df, height=6, aspect=1.5)
plt.title('Boxen Plot: Sale Price by House Style')
plt.xticks(rotation=45)
plt.show()
```

### 8. Customizing & Saving Plots

8.1. Custom Legends, Size, and Annotations

```python
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, hue='CentralAir', size='BedroomAbvGr', sizes=(40, 200))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Living Area vs Sale Price by Air Conditioning and Bedrooms')
plt.tight_layout()
plt.show()
```

8.2. Save Your Plot

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='BedroomAbvGr', y='SalePrice', data=df)
plt.title('Sale Price by Number of Bedrooms')
plt.savefig("saleprice_by_bedrooms.png", dpi=300, bbox_inches='tight')
plt.show()
```

# 🚀 Pro Tips for Beginners

- **Always inspect your data before plotting.**  
  Understanding your dataset's structure and quality ensures meaningful visualizations.

- **Leverage `hue`, `size`, and `style` parameters in Seaborn.**  
  These add additional dimensions to your plots, making them richer and more informative.

- **Adjust figure size for better clarity.**  
  Use `plt.figure(figsize=(width, height))` to control the size and improve readability.

- **Save your best plots for reports or presentations.**  
  Export high-quality images using `plt.savefig()` to document and share your insights.

- **Experiment with different Seaborn themes.**  
  Themes like `whitegrid`, `darkgrid`, and `ticks` provide different aesthetics and can enhance visual appeal.

- **Utilize `FacetGrid` and `catplot` for multi-panel visualizations.**  
  These tools allow you to create complex, faceted plots that reveal patterns across multiple categories.

1. Customizing Color Palettes and Themes

```python
# Built-in theme এবং palette সেট করা
sns.set_theme(style="darkgrid", palette="pastel", font_scale=1.2)

# Custom color palette তৈরি করা
custom_palette = sns.color_palette(["#FF5733", "#3366FF", "#33FF33"])

# Custom palette দিয়ে plot
sns.boxplot(x="day", y="total_bill", data=sns.load_dataset("tips"), palette=custom_palette)
plt.title("Boxplot with Custom Color Palette")
plt.show()
```

2. Complex Multi-Plot Grids (FacetGrid, PairGrid)

```python
# FacetGrid ব্যবহার করে একই ধরনের plot একসাথে দেখানো
g = sns.FacetGrid(sns.load_dataset("tips"), col="time", row="smoker", height=4)
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()

# PairGrid এ কাস্টম ফাংশন ম্যাপ করা
g = sns.PairGrid(sns.load_dataset("iris"))
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_diag(sns.histplot, kde=True)
plt.show()
```

3. Advanced Statistical Visualizations (Estimator, Confidence Intervals, Bootstrapping)

```python
# Categorical aggregation control with estimator (mean, median)
sns.barplot(x="day", y="total_bill", data=sns.load_dataset("tips"), estimator=sum, ci=95)
plt.title("Sum of Total Bill per Day with 95% CI")
plt.show()

# Bootstrapping example (default in seaborn)
sns.pointplot(x="day", y="total_bill", data=sns.load_dataset("tips"), capsize=.2)
plt.title("Pointplot with Bootstrapped Confidence Intervals")
plt.show()
```

4. Annotations and Detailed Axis/Tick Customization

```python
tips = sns.load_dataset("tips")
plt.figure(figsize=(8, 6))
ax = sns.barplot(x="day", y="total_bill", data=tips)

# Annotation যোগ করা
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

# Axis label এবং tick customization
ax.set_xlabel("Day of Week", fontsize=12)
ax.set_ylabel("Total Bill", fontsize=12)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.title("Total Bill by Day with Annotations")
plt.show()
```

5. Handling Large Datasets Efficiently (Subsetting & Sampling)

```python
import pandas as pd
import numpy as np

# বড় dataset তৈরি (উদাহরণ)
large_df = pd.DataFrame({
    "x": np.random.randn(100000),
    "y": np.random.randn(100000),
    "category": np.random.choice(["A", "B", "C"], 100000)
})

# Subset বা sample নেওয়া
sample_df = large_df.sample(n=2000, random_state=42)

# Sample দিয়ে scatterplot
sns.scatterplot(x="x", y="y", hue="category", data=sample_df)
plt.title("Scatterplot on Sampled Large Dataset")
plt.show()
```

6. Integration with Pandas Data Structures and Complex Data Types

```python
# Time series data উদাহরণ
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
ts_df = pd.DataFrame(date_rng, columns=['date'])
ts_df['data'] = np.random.randint(0, 100, size=(len(date_rng)))

# Plotting time series
sns.lineplot(x='date', y='data', data=ts_df)
plt.title("Time Series Plot")
plt.xticks(rotation=45)
plt.show()
```

7. Combining Seaborn with Matplotlib for Ultimate Flexibility

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Seaborn plot on matplotlib axis
sns.scatterplot(x="total_bill", y="tip", data=sns.load_dataset("tips"), ax=ax)

# Matplotlib customization
ax.set_title("Scatterplot with Matplotlib Customization", fontsize=16)
ax.set_xlabel("Total Bill ($)", fontsize=14)
ax.set_ylabel("Tip ($)", fontsize=14)
ax.grid(True)

# Adding horizontal line
ax.axhline(y=5, color='red', linestyle='--', linewidth=1)

plt.show()
```
