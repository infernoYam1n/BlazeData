#### Salary Prediction using Linear Regression

```python
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv("Salary Data.csv")
df.head()
```

#### EDA

```python
plt.scatter(df['Experience Years'],df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

```python
x=df[['Experience Years']]
y=df['Salary']
```

```python
plt.scatter(x.mean(),y.mean(),color='red',marker='*')
```

```python
plt.scatter(x.mean(),y.mean(),color='red',marker='*')

plt.scatter(df['Experience Years'],df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
reg=LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
reg.fit(x,y) 
```

```python
m=reg.coef_
m
```

```python
c=reg.intercept_
c
```

```python
y=m*x+c
y
```

```python
df['predicted_y']=reg.predict(x)
df.head()
```

```python
plt.plot(x,reg.predict(x))
```

```python
plt.plot(x,df['predicted_y'])
plt.scatter(x.mean(),y.mean(),color='red',marker='*')

plt.scatter(df['Experience Years'],df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

#### Cost & Lose

```python
df['loss']=df['Experience Years']-df['predicted_y']
df.head()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse=mean_squared_error(df['Experience Years'],df['predicted_y'])
mse 
```

```python
mae=mean_absolute_error(x,df['predicted_y'])
mae
```

```python
sum(df['loss'])
```

```python
sum(abs(df['loss']))/len(x)  #mean absolute error
```

```python
plt.plot(x,df['predicted_y'])
plt.scatter(x.mean(),y.mean(),color='red')

plt.scatter(df['Experience Years'],df['Salary'])

plt.scatter(df['Experience Years'],df['predicted_y'],color='green')

plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y,reg.predict(x))
```

```python
reg.score(x,y)
```
