```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
x=housing.data
y=housing.target

data=pd.DataFrame(x,columns=housing.feature_names)
data["SalesPrice"]=y
data.head()
```

```python
fig,ax=plt.subplots()
ax.scatter(x=data['MedInc'], y=data['SalesPrice'])
plt.ylabel('SalesPrice', fontsize=13)
plt.xlabel('MedInc', fontsize=13)
plt.title('SalesPrice vs MedInc', fontsize=15)
plt.show()
```

```python
from scipy import stats
from scipy.stats import norm, skew
import seaborn as sns
import matplotlib.pyplot as plt

# Fit the normal distribution
(mu, sigma) = norm.fit(data['SalesPrice'])
print('\n mu = {:.2f} and sigma = {:.2f}'.format(mu, sigma))

# Plot histogram with normal distribution fit
sns.histplot(data['SalesPrice'], kde=False, stat="density", bins=30)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f})'.format(mu, sigma)])
plt.title('SalesPrice distribution')

# Q-Q plot
fig = plt.figure()
res = stats.probplot(data['SalesPrice'], plot=plt)
plt.title('Q-Q plot')
plt.show()
```

```python
data["SalesPrice"]=np.log1p(data["SalesPrice"])
fig,ax=plt.subplots()
ax.scatter(x=data['MedInc'], y=data['SalesPrice'])
plt.ylabel('SalesPrice', fontsize=13)
plt.xlabel('MedInc', fontsize=13)
plt.title('SalesPrice vs MedInc', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,6))
cor=data.corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.PuBu)
plt.show()
```

```python
cor_target=abs(cor["SalesPrice"])
relevant_features=cor_target[cor_target>0.2]

names=[index for index, value in relevant_features.items()]

names.remove("SalesPrice")
print(names)
print(len(names))
```

```python
from sklearn.model_selection import train_test_split
x=data.drop("SalesPrice",axis=1)
y=data["SalesPrice"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictions=lr.predict(x_test)

print("Actual value of the house: ",y_test[0])
print("Predicted value of the house: ",predictions[0])
```

```python
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y_test,predictions)
rmse=np.sqrt(mse)
print(mse)
```
